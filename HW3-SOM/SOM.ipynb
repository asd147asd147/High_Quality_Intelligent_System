# 2. No topological structure

```python
np.random.seed(11)
Weight = np.random.rand(w*h,25).T

No_topological_structure = np.copy(char_set)
Weight_rate = 0.6
N = 100
for _ in range(N):
    for k in range(3):
        for i in range(7):
            dist = np.zeros((25,))
            for j in range(25):
                flat_data = np.copy(np.ravel(No_topological_structure[k][i]))
                dist[j] = np.linalg.norm(flat_data-np.copy(Weight[j]))
            min_index = dist.argmin()
            Weight[min_index] = Weight[min_index]+Weight_rate*(flat_data - Weight[min_index])
        Weight_rate /= 2
print(Weight)
```

```python
import pandas as pd

result = [[] for _ in range(25)]

for k in range(3):
    for i in range(7):
        dist = np.zeros((25,))
        for j in range(25):
            flat_data = np.copy(np.ravel(No_topological_structure[k][i]))
            dist[j] = np.linalg.norm(flat_data-np.copy(Weight[j]))
        min_index = dist.argmin()
        result[min_index].append((k,i))

Units = []
Patterns = []
for i in range(25):
    Pattern = []
    if result[i]:
        Units.append(i+1)
        for j in result[i]:
            Pattern.append(char_list[j[1]]+str(j[0]+1))
    if Pattern:
        Patterns.append(' '.join(Pattern))

df_NTS = pd.DataFrame(data = np.array([Units,Patterns]).T, columns=['Unit','Patterns'])

display(df_NTS,'No Topological Structure')
```

# 3. Linear structure

```python
np.random.seed(11)
Weight = np.random.rand(w*h,25).T

Linear_structure = np.copy(char_set)
Weight_rate = 0.6
N = 100
for _ in range(N):
    for k in range(3):
        for i in range(7):
            dist = np.zeros((25,))
            for j in range(25):
                flat_data = np.copy(np.ravel(Linear_structure[k][i]))
                dist[j] = np.linalg.norm(flat_data-np.copy(Weight[j]))
            min_index = dist.argmin()
            Weight[min_index] = Weight[min_index]+Weight_rate*(flat_data - Weight[min_index])
            if min_index != 1:
                Weight[min_index-1] = Weight[min_index-1]+Weight_rate*(flat_data - Weight[min_index-1])
            if min_index != 24:
                Weight[min_index+1] = Weight[min_index+1]+Weight_rate*(flat_data - Weight[min_index+1])
        Weight_rate /= 2
print(Weight)
```

# 4. Diamond structure

```python
np.random.seed(11)
Weight = np.random.rand(w*h,25).T

Diamond_structure = np.copy(char_set)
Weight_rate = 0.6
N = 100
for _ in range(N):
    for k in range(3):
        for i in range(7):
            dist = np.zeros((25,))
            for j in range(25):
                flat_data = np.copy(np.ravel(Diamond_structure[k][i]))
                dist[j] = np.linalg.norm(flat_data-np.copy(Weight[j]))
            min_index = dist.argmin()
            Weight[min_index] = Weight[min_index]+Weight_rate*(flat_data - Weight[min_index])
            if min_index != 1:
                Weight[min_index-1] = Weight[min_index-1]+Weight_rate*(flat_data - Weight[min_index-1])
            if min_index != 24:
                Weight[min_index+1] = Weight[min_index+1]+Weight_rate*(flat_data - Weight[min_index+1])
            if min_index >= 5:
                Weight[min_index-5] = Weight[min_index-5]+Weight_rate*(flat_data - Weight[min_index-5])
            if min_index <= 19:
                Weight[min_index+5] = Weight[min_index+5]+Weight_rate*(flat_data - Weight[min_index+5])
            
        Weight_rate /= 2
print(Weight)
```
